#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
''' 
accepts the sample data points and population mean and returns standard deviation for the given sample.
'''
def calc_std(col,sample_data_points , mu):
      n= len(sample_data_points)        
      summation =0  
      for i in range (0,n):
         summation += ((sample_data_points[i]-mu)**2)

      std_dev =  (summation /n )**0.5
      return(std_dev) 
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

````python
# calculating standard deviation for all numeric columns
{col : calc_std(col,final_df[col],final_df[col].mean()) for col in (final_df.select_dtypes('int64').columns |
                                                                final_df.select_dtypes('float64').columns)}
````

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels,list_test_type, list_test_year,mean_list ,xlabel ):
    plt.figure(figsize=(12,12))
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    #fig, ax = plt.subplots(nrows=nrows, ncols=2) # You'll want to specify your figsize
    ax= [i for i in range(len(list_of_columns))]
    #ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i] =plt.subplot(nrows,2,i+1)
        ax[i].hist(dataframe[column]) # feel free to add more settings
        #Set titles, labels, etc here for each subplot
        ax[i].set_title(f' {list_test_year[i]} {list_test_type[i]} {xlabel} distribution')
        ax[i].set_xlabel(f"{xlabel} {list_test_year[i]}")
        ax[i].set_ylabel(f"frequency")
        ax[i].vlines(x= mean_list[i],label='mean', ymin =0,ymax=10,color ='y',lw=5 );
        
# credit :create datframe sample code https://www.geeksforgeeks.org/creating-a-dataframe-from-pandas-series/
normalized_percent_scores = { 'percent_act_score_17': final_df['act_compo_17'] *100/31,
          'percent_act_score_18': final_df['act_compo_18'] *100/31 ,
          'percent_sat_score_17': final_df['sat_tot_17'] * 100/1600,
          'percent_sat_score_18': final_df['sat_tot_18'] * 100/1600,
         } 


subplot_histograms(pd.DataFrame(normalized_percent_scores) ,
                  ['percent_act_score_17','percent_act_score_18','percent_sat_score_17','percent_sat_score_18'],
                  ['percent_act_score_17','percent_act_score_18','percent_sat_score_17','percent_sat_score_18'],
                  ['percent_act_score_17','percent_act_score_18','percent_sat_score_17','percent_sat_score_18'],
                  [ 'ACT','ACT','SAT','SAT'],
                  [ '2017','2018','2017','2018'],
                  [ normalized_percent_scores['percent_act_score_17'].mean(),normalized_percent_scores['percent_act_score_18'].mean(),
                   normalized_percent_scores['percent_sat_score_17'].mean(),normalized_percent_scores['percent_sat_score_18'].mean()],
                  'percent score ')

````
